```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
datagen = ImageDataGenerator(rotation_range=40,
                            width_shift_range=0.2,
                            height_shift_range=0.2,
                            shear_range=0.2,
                            zoom_range=0.2,
                            horizontal_flip=True,
                            fill_mode='nearest')
```

```python
img = load_img('C:\\Users\\EXCELL  COMPUTERS\\Programming_Data_Science\\Convolutional_Neural_Network\\Cat.jpg')
```

```python
img
```

```python
x = img_to_array(img)
```

```python
x
```

```python
x.shape
```

```python
x = x.reshape((1,)+ x.shape)
```

```python
x.shape
```

```python
i = 0
for batch in datagen.flow(x , batch_size=1,save_to_dir='preview', save_prefix='cat',save_format='jpeg'):
    i+= 1
    
    if i>20:
        break
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import glob
directory = 'preview'

image_files = glob.glob(f"{directory}/*.jpeg")

rows = 4
columns = 5

fig, axes = plt.subplots(rows, columns, figsize=(12, 10))

# Loop through the image files and display each one in a subplot
for i, ax in enumerate(axes.flat):
    if i < len(image_files):
        img = Image.open(image_files[i])
        ax.imshow(img)
        ax.axis('off')
    else:
        ax.axis('off')  # Hide empty subplots

plt.show()
```
